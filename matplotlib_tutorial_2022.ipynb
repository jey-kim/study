## Functional Plot

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1 ** 2
plt.plot(x_1,y_1)
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.show()
```

## Print Multiple plots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

## Using Figure Objects

```python
fig_1 = plt.figure(figsize=(5,4),dpi=100) #5 inch, 4 inch
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) #bottom left (x,y) and del x and del y
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x   $x^2$')
axes_1.plot(y_1,x_1,label='$x^2$  x')
axes_1.legend(loc=0) #loc = 0 tries to find the best place
#upper right: 1, upper left: 2, lower left: 3, lower right: 4
# or supply a tuple of x & y from lower left

#add figure in a figure
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,y_1,'r')

axes_2.text(0,40,'Message') #
```

## Pie Charts

```python
import random

fig_6 = plt.figure(figsize=(8,5), dpi=100)
axes_6 = fig_6.add_axes([0.1,0.1,0.9,0.9])

types = ['Water','Normal','Flying','Grass','Psychic','Bug',
         'Fire','Poison','Ground','Rock','Fighting','Dark',
         'Steel','Electric','Dragon','Fairy','Ghost','Ice']

poke_num = [133,109,101,98,85,77,68,66,65,60,57,54,53,51,50,50,46,40]

#random color
colors = []
for i in range(len(poke_num)):
    rgb = (random.uniform(0.5,1),random.uniform(0.5,1),random.uniform(0.5,1)) # Dark 0,0.5 ; Bright 0.5,1
    colors.append(rgb)
 
explode = [0] * 18
explode[0] = 0.2 # Explode out the first
wedges, texts, autotexts =plt.pie(poke_num, explode=explode, 
                                  labels=types, colors=colors, 
                                  autopct='%1.0f%%',shadow=False,
                                 startangle=142, textprops=dict(color="k"))
#autopct: auto percentage

plt.legend(wedges, types, loc='right', bbox_to_anchor=(1,0,0.35,1))
```

## Timeseries

```python
import datetime as dt
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday, nearest_workday, \
    USMartinLutherKingJr, USPresidentsDay, GoodFriday, USMemorialDay, \
    USLaborDay, USThanksgivingDay

class USTradingCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('NewYearsDay', month=1, day=1),
        USMartinLutherKingJr,
        USPresidentsDay,
        GoodFriday,
        USMemorialDay,
        Holiday('USIndependenceDay', month=7, day=4, observance=nearest_workday),
        USLaborDay,
        USThanksgivingDay,
        Holiday('Christmas', month=12, day=25, observance=nearest_workday)
        ]
```

```python
goog_data = pd.read_csv('GOOG.csv') # Yahoo Finance > search GOOG > Historical Data > Download
goog_data_np = goog_data.to_numpy()

goog_cp = goog_data_np[:,4]
date_arr_np = goog_data_np[:,0]

cal = USTradingCalendar()
holidays = cal.holidays(start=goog_data['Date'].min(), end=goog_data['Date'].max())
date_arr = pd.bdate_range(start=goog_data['Date'].min(), end=goog_data['Date'].max(), freq='C', holidays=holidays) # freq='C'
date_arr_np = date_arr.to_numpy()

fig_7=plt.figure(figsize=(8,5))
axes_7 = fig_7.add_axes([0.1,0.1,0.9,0.9])
plt.plot(date_arr_np,goog_cp)
```

```python
def get_z(x, y):
    return np.sin(np.sqrt(x**2+y**2))

fig_10 = plt.figure(figsize=(8,5), dpi=100)
axes_10 = fig_10.add_axes([0.1,0.1,0.9,0.9], projection='3d')
x_4 = np.linspace(-6,6,30)
y_4 = np.linspace(-6,6,30)

x_4, y_4 = np.meshgrid(x_4,y_4)
z_4 = get_z(x_4,y_4)

axes_10.view_init(45,55) #viewing angle 45 degree, rotate 55 degree
axes_10.contour3D(x_4,y_4,z_4, 20, cmap='Blues')  # 20 contour lines
```

```python
fig_11 = plt.figure(figsize=(8,5), dpi=100)
axes_11 = fig_11.add_axes([0.1,0.1,0.9,0.9], projection='3d')


axes_11.view_init(45,55) #viewing angle 45 degree, rotate 55 degree
axes_11.set_xlabel('x')
axes_11.set_ylabel('y')
axes_11.set_zlabel('z')
axes_11.plot_wireframe(x_4,y_4,z_4, color='blue')  # 20 contour lines
```

```python
fig_12 = plt.figure(figsize=(8,5), dpi=100)
axes_12 = fig_12.add_axes([0.1,0.1,0.9,0.9], projection='3d')


axes_12.view_init(45,55) #viewing angle 45 degree, rotate 55 degree
axes_12.set_xlabel('x')
axes_12.set_ylabel('y')
axes_12.set_zlabel('z')
axes_12.plot_surface(x_4,y_4,z_4, rstride=1,cstride=1,cmap='Blues',edgecolor='none') 
```
