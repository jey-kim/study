## every signal has even and add parts

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate a random signal
x = np.random.rand(9) -.5
```

```python
# Plot the random signal placing the middle value at 0
plt.stem(np.arange(-4,5),x)
plt.show()
```

```python
negx=x[::-1] #flip the x
```

```python
plt.stem(np.arange(-4,5),negx)
plt.show()
```

```python
evx = (x + negx) / 2 #even part of x
odx = (x - negx) / 2 #add part of x
```

```python
plt.stem(np.arange(-4,5),evx)
plt.show()
```

```python
plt.stem(np.arange(-4,5),odx)
plt.show()
```

```python
q = evx + odx #sum should be the same
```

```python
plt.stem(np.arange(-4,5),q)
plt.show()
```
