```python
plt.figure(figsize=(8,5))

sns.stripplot(x = 'day', y='total_bill', data=df_tips, jitter=True, hue='sex', dodge=True) #draw a scatter plot
#dodge: separate men and women
#jitter: spread your data
```

PairGrid

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species') #color based on species
#iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris_g = sns.PairGrid(iris, hue='species') #color based on species
#iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g = sns.PairGrid(iris, hue='species'
                      , x_vars=["sepal_length", "sepal_width"]
                      , y_vars=["petal_length", "petal_width"])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

FacetGrid

```python
tips_fg = sns.FacetGrid(df_tips, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
```

```python
tips_fg = sns.FacetGrid(df_tips, col='time', row='smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(df_tips, col='time', hue='smoker', height=4, aspect=1.3 
                        , col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

```python
kws = dict(s=50, linewidth=.5, edgecolor = 'w') # kws = keywords
tips_fg=sns.FacetGrid(df_tips, col='sex', hue='smoker', height=4, aspect=1.3,
                      hue_order=['Yes', 'No'],
                      hue_kws=dict(marker=['^','v']))


tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

```python
df_att = sns.load_dataset('attention')
att_fg = sns.FacetGrid(df_att, col='subject' , col_wrap=5, height=1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
```
