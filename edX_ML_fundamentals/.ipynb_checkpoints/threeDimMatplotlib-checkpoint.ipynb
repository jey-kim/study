```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import multivariate_normal
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14)
```

```python
mu = np.array([10,10])
sigma = np.array([[1, 1], [1, 3]])
```

```python
rv = multivariate_normal(mean=mu, cov=sigma)
```

```python
x = np.arange(1,21)
y = np.arange(1,21)

xmin = min(x) - 1
xmax = max(x) + 1
ymin = min(y) - 1
ymax = max(y) + 1
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
grid_spacing = 0.05
xx1, xx2 = np.meshgrid(np.arange(xmin, xmax, grid_spacing), np.arange(ymin, ymax, grid_spacing))
grid = np.c_[xx1.ravel(), xx2.ravel()]
Z = np.array([rv.logpdf(pt) for pt in grid])
Z = Z.reshape(xx1.shape)
plt.pcolormesh(xx1, xx2, Z, cmap=plt.cm.PRGn, vmin=-30, vmax=-3)
```

```python
np.min(Z)
np.max(Z)
```
