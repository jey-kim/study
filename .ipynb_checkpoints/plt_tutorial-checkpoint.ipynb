```python
# evenly sampled time at 200ms intervals
t=np.arange(0., 5., 0.2)

#red dashed line; blue squares; green triangles
plt.plot(t, t, 'r--', \
         t, t**2, 'bs',\
         t, t**3, 'g^') 
plt.show()
```


# Plotting with keyword strings

```python
dictionary = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}

dictionary['b'] = dictionary['a'] + 10 * np.random.randn(50)
dictionary['d'] = np.abs(dictionary['d']) * 100
```

```python
plt.scatter('a','b', \
            c='c', s='d',data=dictionary)

#c=color!
#d=size!

plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group a','group b','group c']
values = [1,10,100]

plt.figure(figsize=(9,3))


plt.subplot(131)
plt.title("bar")
plt.bar(names, values)
plt.subplot(132)
#plt.title("scatter")
plt.scatter(names, values)
plt.subplot(133)
plt.title("plot")
plt.plot(names, values)
plt.suptitle("Categorical Plotting") #title for all
plt.show()
```

```python
#IQ
mu = 100  # mean of distribution 
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)
num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()

plt.savefig('IQ.pdf',bbox_inches='tight')
plt.show()
```
