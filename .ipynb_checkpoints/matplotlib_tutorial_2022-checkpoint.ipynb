## Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
## make all of your plots show up in the notebook
%matplotlib inline  

%reload_ext autoreload
%autoreload 2
```

## Functional Plot

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1 ** 2
plt.plot(x_1,y_1)
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.show()
```

## Print Multiple plots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

## Using Figure Objects

```python
fig_1 = plt.figure(figsize=(5,4),dpi=100) #5 inch, 4 inch
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) #bottom left (x,y) and del x and del y
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x   $x^2$')
axes_1.plot(y_1,x_1,label='$x^2$  x')
axes_1.legend(loc=0) #loc = 0 tries to find the best place
#upper right: 1, upper left: 2, lower left: 3, lower right: 4
# or supply a tuple of x & y from lower left

#add figure in a figure
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,y_1,'r')

axes_2.text(0,40,'Message') #
```

## SubPlots

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout() # no overlap
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[0].set_ylabel('Haha')
axes_2[1].plot(x_1,y_1)
axes_2[2].plot(x_1,y_1,'r*')
```

## Appearance Options

```python
## Default colors (b,r,g,c,m,y,k,w)
## color = "0.75" creates a 75% gray  (0=black, 1=white)
## You can use hexcodes color="#eeefff"
## You can use color names found next like this color="burlywood"
## https://en.wikipedia.org/wiki/Web_colors



fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1]) #take up all the space
axes_3.plot(x_1,y_1, color = 'navy', alpha=.620, lw=2, ls='-.', marker='o', markersize=7,
           markerfacecolor='y', markeredgecolor='k', markeredgewidth=1) 
## linewidth lw=1 is default
## transparency alpha
## linestyle ls

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

axes_3.grid(True, color='0.1', dashes=(10,2,1,2)) #5pts, 2spaces 1pt 2 spaces
axes_3.set_facecolor('#FAEBD7')
```

## Working with Pandas DataFrame

```python
data={'Temperature': [37,40,49,61,72,79,83,81,75,64,53,42],
      'Sales': [292,228,324,376,440,496,536,556,496,412,324,320]}
```

```python
ics_df = pd.DataFrame(data)
ics_df = ics_df.sort_values(by='Temperature')
ics_df

np_arr = ics_df.values
x_2 = np_arr[:,0]
y_2 = np_arr[:,1]
```

```python
fig_4 = plt.figure(figsize=(6,4))
axes_4 = fig_4.add_axes([0,0,1,1])
axes_4.set_xlabel('Temperature')
axes_4.set_ylabel('Ice Cream Sales')
axes_4.set_title('Ice Cream Sales vs. Temperature')
axes_4.plot(x_2,y_2)

axes_4.annotate('Good Month', xy=(83,536), xytext=(60,520), 
                arrowprops=dict(facecolor='k',shrink=0.05))

plt.bar(x_2,y_2)
```
