```python

import pygmt
import numpy as np
```

```python
#info = np.loadtxt('seismogram_input.txt')

info = np.array([[126.6239, 37.4776, 29.183, 1.752],
[126.9667, 37.5667, 28., 1.745],   
[127.363808, 36.37719, 30.763, 1.712],  
[129.112503, 35.2486, 31.806, 1.749],
[127.9748, 36.873001, 30.669, 1.737],
[127.3964, 35.421299, 22.657, 1.523],
[128.252502, 38.2686, 28.689, 1.773],
[126.917099, 37.4939, 28.324, 1.735]])
```

```python
longitude = info[:,0]
latitude = info[:,1]
mohodepth = info[:,2]
vpOVERvs = info[:,3]
print(len(mohodepth))
```

```python
region = [
    longitude.min() - 1,
    longitude.max() + 1,
    latitude.min() - 1,
    latitude.max() + 1,
]
print(region)
```

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="black", water="grey")
pygmt.makecpt(cmap="viridis", series=[mohodepth.min(), mohodepth.max()])
```

```python
fig.plot(x=longitude, y=latitude)
fig.colorbar(frame='af+l"Depth (km)"')
```

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="black", water="grey")
pygmt.makecpt(cmap="viridis", series=[mohodepth.min(), mohodepth.max()])
fig.plot(
    x=longitude,
    y=latitude,
    color=mohodepth,
    cmap=True,
    style="cc",
    pen="black",
)
fig.colorbar(frame='af+l"Depth (km)"')
fig.show()
```
